### K-Means

```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
x = iris.data[:, :4]
print(x.shape)
# print(x)
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,1,1)
plt.scatter(x[:,0], x[:,1],c="red", marker='o', label='datas')
plt.xlabel("datas")
plt.legend(loc=0)
```

```python
# 操作名稱 = 分群算法(參數)
estimator = KMeans(n_clusters=3)

# 操作名稱.fit(數據)
estimator.fit(x)

# labels_ => 訓練完的結果
label_pred = estimator.labels_

label_pred
```

```python
plt.subplot(2, 1, 2)
x0 = x[label_pred == 0]
x1 = x[label_pred == 1]
x2 = x[label_pred == 2]
plt.scatter(x0[:, 0], x0[:, 1], c='red', marker='o', label='labele')
plt.scatter(x1[:, 0], x1[:, 1], c='green', marker='*', label='labell')
plt.scatter(x2[:, 0], x2[:, 1], c='blue', marker='+', label='label2')
plt.xlabel('K-means')
plt.legend(loc=0)
```

* 上圖有些點看起來是重疊的 => 因為實際上應該是三個維度

  => 但因為不好畫，我們畫二維

  => 所以將圖看成三維，實際上點是分開的

```python
# 操作名稱 = 分群算法(參數)
estimator = KMeans(n_clusters=4)

# 操作名稱.fit(數據)
estimator.fit(x)

# labels_ => 訓練完的結果
label_pred = estimator.labels_
```

```python
plt.subplot(2, 1, 2)
x0 = x[label_pred == 0]
x1 = x[label_pred == 1]
x2 = x[label_pred == 2]
x3 = x[label_pred == 3]
plt.scatter(x0[:, 0], x0[:, 1], c='red', marker='o', label='labele')
plt.scatter(x1[:, 0], x1[:, 1], c='green', marker='*', label='labell')
plt.scatter(x2[:, 0], x2[:, 1], c='blue', marker='+', label='label2')
plt.scatter(x3[:, 0], x3[:, 1], c='orange', marker='^', label='label3')
plt.xlabel('K-means')
plt.legend(loc=0)
```
